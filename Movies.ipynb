## NLP Text Preprocessing

```python
import nltk 
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text: str) -> str:
    # Convert the string to lowercase 
    text = str(text).lower()

    # Remove all punctuations
    text = re.sub(r'[^\w\s]', '', text)

    # Remove all stopwords
    words = text.split()
    words = [word for word in words if word not in stop_words]

    # Lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]

    # Return the result as a string
    return ' '.join(words)
```

```python
# Apply the function to tags
df_model['tags'] = df_model['tags'].apply(preprocess_text)
```

```python
df_model.head()
```

## Text Vectorization

```python
# Reset the indices 
df_model = df_model.reset_index(drop=True)
```

```python
indices = pd.Series(df_model.index, index=df_model['title']).drop_duplicates()
indices
```

#### Applying TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,
                        ngram_range=(1, 2),
                        stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df_model['tags'])
```

```python
tfidf_matrix
```

## Cosine Similiarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title: str, n=10):
    if title not in indices:
        return ['❌ Movie not found']

    index = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[index], tfidf_matrix).flatten()
    similiar_idx = sim_score.argsort()[::-1][1:n+1]
    return df_model['title'].iloc[similiar_idx]
    
```

```python
recommend('Avatar')
```

## Save the TF-IDF Matrix and the Dataframe

```python
import pickle as pkl
import os

os.mkdir('models/')

pkl.dump(tfidf_matrix, open('models/tfidf_matrix.pkl', 'wb'))

pkl.dump(indices, open('models/indices.pkl', 'wb'))

df_model.to_pickle('models/df_model.pkl')

pkl.dump(tfidf, open('models/tfidf.pkl', 'wb'))
```
